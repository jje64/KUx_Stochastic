```python
%matplotlib inline
NSTEP = 1000 # number of random steps
plt.xlabel("number of steps") # name of x axis
plt.ylabel("position") # name of y axis
for nseed in range(10):# generate 10 random walks
    # initialize random step generator with different nseeds
    np.random.seed(nseed) 
    # generate random sequencies of NSTEP +1/-1 steps
    step = np.random.choice([-1,1],NSTEP)
    #print(step)
    # calculate position of random walk at each step
    position = np.cumsum(step)
    plt.plot(position) # plot position(step) vs. step
```

## The Euler Method for Numerical Integration

### The Euler Method
- Let's consider the differential equation of the form:
$$ $$ $$\frac{dy\left(t\right)}{dt}=f\left(y\left(t\right),t\right)$$ $$ $$   
  The general solution of this differential equation, using the __initial condition__ $y_0=y\left(t_0\right)$ will be:
  $$ $$ $$y\left(t\right)=y_0+\int_{t_0}^tf\left(y\left(t'\right),t'\right)dt'$$ $$ $$
- To solve a differential equation with the __Euler Method__, we will have to define the following __discrite variables__:
$$ $$ $$\begin{cases}t_i\equiv t_0+i\Delta t\\ y_i\equiv y\left(t_i\right) \\ f_i\equiv f\left(y\left(t_i\right),t_i\right)\end{cases}$$ $$ $$
- Let's now consider the integration of the differential equation in the interval $\left[t_i,t_{i+1}\right]$. Then we will have that:
$$ $$ $$y\left(t_{i+1}\right)=y\left(t_i\right)+\int_{t_i}^{t_{i+1}}f\left(y\left(t'\right),t'\right)dt'$$ $$ $$
$$y_{i+1}=y_i+\int_{t_i}^{t_i+\Delta t}f\left(y\left(t'\right),t'\right)dt'$$  
  Let's change the integration variable to $\mu=t'-t_i$, thus we will get:
  $$ $$ $$y_{i+1}=y_i+\int_0^{\Delta t}f\left(y\left(\mu+t_i\right),\mu+t_i\right)d\mu \tag{1}\label{1}$$
- From the previous equation, we express the integrand as its __Taylor Series__:
$$ $$ $$f\left(y\left(\mu+t_i\right),\mu+t_i\right)=f_i+\mathcal{O}(\mu)+\mathcal{O}(\mu^2)+\cdots$$   
  By replacing the previous expansion in equation $\ref{1}$, we get:
  $$ $$ $$y_{i+1}=y_i+\int_0^{\Delta t}\left[f_i+\mathcal{O}(\mu)+\mathcal{O}(\mu^2)+\cdots\right]d\mu$$   
  Thus, after integration we get:
  $$ $$ $$y_{i+1}=y_i+f_i\Delta t+\mathcal{O}(\Delta t^2)+\mathcal{O}(\Delta t^3)+\cdots$$   
  Thus for $\Delta t\ll 1$, we can get an approximated value for $y_{i+1}$ using only the first-order term:  
  $$ $$ $$y_{i+1}\approx y_i+f_i\Delta t$$ $$ $$     
- Thus, with this method, we will have an __local truncation error__ of the order $\mathcal{O}(\Delta t^2)$
- It is stated that the __global truncation error__ of the order $\mathcal{O}(\Delta t)$. $\color{red}{\text{Come back to prove it}}$
